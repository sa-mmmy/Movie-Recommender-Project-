```python
#convert jason like strings to python object
```

```python
movies_metadata['genres']=movies_metadata['genres'].fillna('[]').apply(literal_eval).apply(lambda x : [d['name'] for d in x] )
```

```python
movies_metadata['genres']
```

```python
def clean_keywords(x):
    # Case 1: String
    if isinstance(x, str):
        try:
            data = literal_eval(x)
        except:
            return []
    else:
        data = x
    
    # Case 2: Already list of dicts
    if isinstance(data, list):
        names = []
        for d in data:
            if isinstance(d, dict) and "name" in d:
                names.append(d["name"])
            elif isinstance(d, str):  # Case 3: already list of strings
                names.append(d)
        return names
    
    return []
```

```python
keywords['keywords'] = keywords['keywords'].fillna('[]').apply(clean_keywords)
```

```python
keywords['keywords']
```

```python
credits.info()
```

```python
# Extract only the Director from crew
credits['crew'] = credits['crew'].fillna('[]') \
    .apply(literal_eval) \
    .apply(lambda x: [d['name'] for d in x if d['job']=="Director"])
```

```python
credits['crew']
```

```python
# Extract only the Director from crew
credits['cast'] = credits['cast'].fillna('[]') \
    .apply(literal_eval) \
    .apply(lambda x: [d['name'] for d in x ])
```

```python
credits['cast']
```

```python
#now we build the soup
```

```python
# ensure id is numeric
```

```python
movies_metadata['id'] = pd.to_numeric(movies_metadata['id'], errors='coerce')
keywords['id'] = pd.to_numeric(keywords['id'], errors='coerce')
credits['id'] = pd.to_numeric(credits['id'], errors='coerce')
```

```python
# Merge
movies = movies_metadata.merge(keywords, on='id').merge(credits, on='id')
```

```python
def create_soup(x):
    return ' '.join(x['genres']) + ' ' + ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + ' '.join(x['crew'])
```

```python
movies_metadata['soup'] = movies.apply(create_soup, axis=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Vectorize the soup
count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(movies_metadata['soup'])

# Compute cosine similarity
cosine_sim = cosine_similarity(count_matrix, count_matrix)
```

```python
# Map indices
indices = pd.Series(movies.index, index=movies['title']).drop_duplicates()

def get_recommendations(title, cosine_sim=cosine_sim):
    if title not in indices:
        return f"Movie '{title}' not found in dataset."
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]  # top 10
    movie_indices = [i[0] for i in sim_scores]
    return movies['title'].iloc[movie_indices]

```

```python
print(get_recommendations("The Matrix"))
```

```python
print(get_recommendations("Tom Hanks"))
```
